# BROADCASTING ARRAYS , VECTORS , MATRIX , TENSORS

```python
L= [10,20,30,40,50] # broadcasting array-1D, VECTRORS
import numpy as np
ADD= np.array(L)
ADD
```

```python
# broadcasting in numpy allows operations on arrays of different shapes.
# Here, we add 10 to each element of the array 'ADD'.
ADD + 10
```

```python
# bROADCASTING IN 2D ARRAY 
MAT = np.arange(1,26).reshape(5,5)
MAT
```

```python
#broadcasting 2D array and ADD 10 to each element
MAT + 10
```
