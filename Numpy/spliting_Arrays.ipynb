# SPLITING ARRAYS
## SPLITING 2D MATRIX

#### HORIZONTAL SPLIT

```python
a = np.arange(16).reshape(4,4)
a # Original array
```

```python
hs = np.hsplit(a, 2) # Split the array into 2 equal parts horizontally
hs
```

```python
hs[0] # First part after horizontal split - index[0]
```

```python
hs[1] # Second part after horizontal split - index[1]
```

```python
hs2 = np.hsplit(a, [1]) # Split the array at column index 2
hs2
```

```python
# if we do it like this hs2 = np.hsplit(a, 1)
# This will split the array into 1 part, which is the original array itself
hs3 = np.hsplit(a, 1) 
hs3 
```

```python
hs4 = np.hsplit(a, 3) # Split the array into 3 equal parts horizontally , but this will give error 
# because 4 columns cannot be evenly split into 3 parts
hs4
```

```python
hs5 = np.hsplit(a, [1,3]) # Split the array at column indices 1 and 3
hs5 # Resulting parts after split at indices 1 and 3
```

```python
hs6 = np.hsplit(a, [0,2,4]) # Split the array at column indices 0, 2, and 4
hs6 # Resulting parts after split at indices 0, 2, and 4
```

```python
hs7 = np.hsplit(a, [0,1,4]) # Split the array at column indices 0, 1, and 4
hs7 # Resulting parts after split at indices 0, 1, and 4
```

```python
hs8 = np.hsplit(a, [0,3,4]) # Split the array at column indices 0, 3, and 4
hs8 # Resulting parts after split at indices 0, 3, and 4
```

```python
hs9 = np.hsplit(a, [2,4]) # Split the array at column indices 2 and 4
hs9 # Resulting parts after split at indices 2 and 4
```

```python
hs10 = np.hsplit(a, [1,2,3]) # Split the array at column indices 1, 2, and 3` 
hs10 # Resulting parts after split at indices 1, 2, and 3
```

```python
# hs10 a different in a way that it splits the array into 4 parts
hs11 = np.hsplit(a, [0,1,2,3,4])
hs11 # Resulting parts after split at indices 0, 1, 2, 3, and 4
# OR
hs12 = np.hsplit(a, 4) # Split the array into 4 equal parts horizontally
hs12 # Resulting 4 equal parts after horizontal split
```

```python
# for clear understanding of hs12 
hs12
```

```python
part1 = hs12[0] # First part
part1
```

```python
part2 = hs12[1] # Second part
part2
```

```python
part3 = hs12[2] # Third part
part3
```

```python
part4 = hs12[3] # Fourth part
part4
```

# VERTICAL SPLIT

```python
b= np.arange(36).reshape(6,6)
b # Original array
```

## Vertical spliting on the basic of Parts

```python
vs = np.vsplit(b, 3) # Split the array into 3 equal parts vertically
vs # Resulting 3 equal parts after vertical split
```

```python
vs1 = np.vsplit(b, 6) # Split the array into 6 equal parts vertically
vs1 # Resulting 6 equal parts after vertical split
```

```python
vs2 = np.vsplit(b,2) # Split the array into 2 equal parts vertically
vs2 # Resulting 2 equal parts after vertical split
```

## ERROR 
#### Caused due to unequal division 

```python
vs3 = np.vsplit(b,4) # Split the array into 4 equal parts vertically , but this will give error
# because 6 rows cannot be evenly split into 4 parts    
vs3
```

## Spliting Vertically on the basic of Index

```python
vs4 = np.vsplit(b,[1]) # Split the array at row index 1
vs4 # Resulting parts after split at index 1
```

```python
vs5 = np.vsplit(b,[2]) # Split the array at row index 2
vs5 # Resulting parts after split at index 2
```

```python
vs6 = np.vsplit(b,[3]) # Split the array at row index 3
vs6 # Resulting parts after split at index 
# Equivalent to np.vsplit(b, 2)
```

```python
vs7 = np.vsplit(b,[4]) # Split the array at row index 4
vs7 # Resulting parts after split at index 4
```

```python
vs8 = np.vsplit(b,[5]) # Split the array at row index 5
vs8 # Resulting parts after split at index 5
```

```python
vs9 = np.vsplit(b,[6])
vs9 # Resulting parts after split at index 6
```

```python
vs10 = np.vsplit(b,[1,3,5]) # Split the array at row indices 1, 3, and 5
vs10 # Resulting parts after split at indices 1, 3, and 5
```

```python
vs11 = np.vsplit(b,[0,2,4,6]) # Split the array at row indices 0, 2, 4, and 6
vs11 # Resulting parts after split at indices 0, 2, 4,  and 6
```

```python
vs12 = np.vsplit(b, [2,4,6]) 
vs12 # Resulting parts after split at indices 2, 4, and 6
```
