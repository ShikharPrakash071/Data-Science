# Data Frames

## Collection of Series is called Data Frames

## 1.  Syntax For DataFrames Using Dictionary
## pd.DataFrames(data)

## 2.  Syntax For DataFrames Using List
## pd.DataFrames(data, column=headers) 

```python
import numpy as np
import pandas as pd
```

## DataFrames In Pandas Using Dictionary With a Single Series

```python
my_list = {1:[11],2:[12],3:[13],4:[14]}
df1 = pd.DataFrame(my_list)
df1
```

## Data Frames Using Dictionary with Multiple Key : Value Pairs (Multiple Series)

```python
data = {
    'Name': ['Tom', 'nick', 'krish', 'jack'],
    'Age': [20, 21, 19, 18],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'Salary': [70000, 80000, 65000, 72000]
    }
df2 = pd.DataFrame(data)
df2
```

## DataFrame in Pandas using List , without giving any Column (HEADERS) values

```python
dlist = [
    ['john', 28, 'New York', 65000],
    ['anna', 22, 'London', 72000],
    ['peter', 34, 'Berlin', 52000]
]

df3 = pd.DataFrame(dlist)
df3
```

## Using Custom Columns (HEADERS)

```python
df4 = pd.DataFrame(dlist, columns=['Name', 'Age', 'City', 'Salary'])
df4
```
